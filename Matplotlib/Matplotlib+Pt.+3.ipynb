### Plot appearance

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='orange')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00') #RGB HEX CODE
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',linewidth=3) #RGB HEX CODE
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',linewidth=3,alpha=0.5) #RGB HEX CODE
#alpha for transparency
#can use linewidth or lw
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',linewidth=3,linestyle='--') #RGB HEX CODE
#alpha for transparency
#can use linewidth or lw
#can use linestyle or ls
```

### Control over axis appearance and plot range

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',linewidth=2,ls='--') #RGB HEX CODE
#alpha for transparency
#can use linewidth or lw

ax.set_xlim([0,1]) #lower bound and upper bound
ax.set_ylim([0,2])
```
